## Policy gradient encoding

Essai d'apprentissage de l'espace latent en utilisant le principe du policy gradient.

structure d'auto-encodeur : z --> u --> z
u est un code latent arbitraire 2D

encodage : z --> u
decodage : u --> z

Encodage continu de type VAE
L'apprentissage du code repose sur la reussite du decodage u --> z mesuré par le reward r (-1 ou 1) selon la reussite (ou non) du decodage + une "consistency loss" implementee par la divergence KL

### Gradient discontinuity
!! pour que le policy gradient fonctionne, il faut "detacher" le sample u_out produit à partir de mu et epsilon => pas de continuité du gradient à travers le code comme dans les VAE.
Si on ne detache pas u_out, le gradient est nul sur les poids de mu...

La validité du code en sortie est mesurée par le multinomial Cross Entropy loss (standard)

### Conclusion

Le policy gradient permet bien de produire un code, avec une bonne couverture de l'espace grace à la contrainte KL.
La discontinuité du gradient permet en théorie d'utiliser n'importe quelle distribution (derivable) sur l'espace latent.

Le "PG encoding" est donc une alternative aux VAE / AAE **MAIS** il faut un critere de renforcement "net" : ici un renforcement binaire +1/-1/ Pas sût que ca marche avec r = BCE loss sur la reconstruction (à essayer... BCE_LOSS > batch average?)

La conclusion principale ici reste qu'il est possible d'apprendre un code avec un reward extrinseque. La combinaison d'un reward extrinseque et d'un reward "local consistency" doit donc permettre d'apprendre un code "parsimonieux"/invariant qui **inclut** l'information de recompense : **reward-aware** encoding : **ce qui n'apporte pas de reward n'est pas encodé **

Le choix de l'action repose alors sur la maximisation de la consistance qui est aussi la maximisation du reward. **le contrôle peut etre appris par optimisation locale (Bandit) ou par predictive control**. 
==> ne sont produites que des reponses motrices susceptibles d'apporter une recompense. / un état z est une configuration de l'environnement dans laquelle une reponse appropriée spécifique doit etre apportée.

```python
%matplotlib inline
import torch

from Where_PG_classif import couples_gen
from scipy.stats import multivariate_normal

import numpy as np
import torch.nn.functional as F


N_test = 10

for _ in range(N_test):
    '''i_offset = np.random.randint(args.size)
    j_offset = np.random.randint(args.size)'''
    mid = np.int(ml.args.size / 2)
    draw = np.random.multivariate_normal((0,0),((1,0),(0,1)))
    i_offset = min(max(-mid, np.int(draw[0] * mid / 3)), mid)
    j_offset = min(max(-mid, np.int(draw[1] * mid / 3)), mid)
    print(i_offset, j_offset)
    data, label = next(iter(ml.dataset.data_loader))
    print (data.shape)  
    idx = 0
    data_full, label_full, pos_full = couples_gen(data[idx, 0, :, :], label[idx], i_offset, j_offset, size=args.size, contrast=1.)
    label_full = np.array(label_full, dtype = 'double')
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots(1, 4, figsize=((15, 5)))
    #ax[0].imshow(a_data)
    ax[0].imshow(data_full)
    ax[0].set_title('image input')
   
    mu_output, logvar_output, u_output, z_hat_logit_output = ml.model(torch.from_numpy(data_full[None, None, :, :]), 
                                                                      torch.from_numpy(label_full[None, :]),
                                                                      torch.from_numpy(pos_full[None, :]))
    #ax[2].imshow(F.sigmoid(output[0, 0, :, :]).detach().numpy())
    #plt.title("("+str(i_offset)+","+str(j_offset)+"), label = "+str(label[idx].item()))
    rep = torch.argmax(z_hat_logit_output).detach().numpy()

    #ig, ax = plt.subplots(1, 3, figsize=((13, 5)))
    #ax[0].plot(a_data)
    ax[1].plot(j_offset, i_offset, 'r+', ms = 15)
    ax[1].set_xlim([-128, 128])
    ax[1].set_ylim([128, -128])
    ax[1].xaxis.grid(True)
    ax[1].yaxis.grid(True)
    ax[1].set_title('actual offset')
    
    u_offset = u_output.detach().numpy()
    mu_offset = mu_output.detach().numpy()
    logvar_offset = logvar_output.detach().numpy()
    DIM1 = 0
    DIM2 = 1
    axe_x, axe_y = np.mgrid[-4:4:.01, -4:4:.01]
    q = multivariate_normal(mu_offset[0], np.exp(logvar_offset[0]) * np.eye(2))
    pos = np.empty(axe_x.shape + (2,))
    pos[:, :, DIM1] = axe_x; pos[:, :, DIM2] = axe_y

    print(mu_offset, u_offset)
    ax[2].contourf(axe_x, axe_y, q.pdf(pos))
    ax[2].plot(u_offset[0][0], u_offset[0][1], 'r+', ms = 15)
    ax[2].xaxis.grid(True)
    ax[2].set_xlim([-4, 4])
    ax[2].set_ylim([-4, 4])
    ax[2].yaxis.grid(True)
    ax[2].set_title('offset PG encoding')    
    ax[2].axis('equal')
    
    
    ax[3].bar(np.arange(10), F.softmax(z_hat_logit_output[0,:]).detach().numpy())
    ax[3].set_ylim([0,1])
    ax[3].set_title("pos = ("+str(i_offset)+","+str(j_offset)+"), label = "+str(label[idx].item())+', rep = '+str(rep))

    plt.show()    
```
